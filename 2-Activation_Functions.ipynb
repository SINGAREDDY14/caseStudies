```python
from scipy.stats import zscore

df_scaled= df_train.apply(zscore)
```

```python
X=df_scaled
Y= df.Signal_Strength
X_train, X_test, y_train, y_test= train_test_split(X, Y , test_size=0.35, random_state=1)
```

## PART-1 Build a Neural Network Regressor

```python
## Training Model-1 for Neural Network Regressor
def regressor():
    model = Sequential()
    model.add(Dense(200, input_shape = (11, )))
    model.add(Activation('relu'))
    model.add(Dense(100, input_shape = (11, )))
    model.add(Activation('relu'))
    model.add(Dense(1))
    sgd = optimizers.SGD(lr = 0.001)
   
    model.compile(optimizer = sgd, loss = 'mse')
    return model
```

```python
model= regressor()
reg = model.fit(X_train, y_train, batch_size=200,  epochs = 100, verbose = 1, validation_split=0.3)
```

```python
## Training Model-2 for Neural Network Regressor
def sigmoid():
  model = Sequential()
  model.add(Dense(8, input_shape = (11, )))
  model.add(BatchNormalization())
  model.add(Activation('sigmoid'))
  model.add(BatchNormalization())
  model.add(Dense(1))
  sgd = optimizers.SGD(lr = 0.001)
   
  model.compile(optimizer = sgd, loss = 'mse')
  return model
```

```python

model_two= sigmoid()
relu= model_two.fit(X_train, y_train, batch_size=200,  epochs = 100, verbose = 1, validation_split=0.3)
```

```python
## Training Model-3 for Neural Network Regressor
def tanh():
    model = Sequential()
    model.add(Dense(7, input_shape = (11, )))
    model.add(BatchNormalization())
    model.add(Activation('tanh'))
    model.add(Dense(1))
    sgd = optimizers.Adam(lr = 0.01)
   
    model.compile(optimizer = sgd, loss = 'mse')
    return model
```

```python
model_three= tanh()
relu_add= model_three.fit(X_train, y_train,  epochs = 100, verbose = 1,validation_split=0.3)
```

From the above three Neural Network for Regressor, it seems quite clear that the last model, model_three has performed the best. 

```python
## Pickling Neural Network Regressor
model_json = model_three.to_json()
with open("model_three.json", "w") as json_file:
    json_file.write(model_json)
```

```python
model_three.save_weights("model_three.h5")
print("Saved model to disk")
```

```python
## Reading a Pickled Neural Network Regressor
json_file = open('model_three.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
```

```python
adam= optimizers.Adam(lr=0.01)
loaded_model.compile(loss='mse', optimizer='adam')
```

```python
loaded_model.load_weights("model_three.h5")
```

```python
## Fitting the pickled model to test data
loaded_model.fit(X_test, y_test, epochs=100, batch_size=100, validation_split=0.35 )
```

## PART II- Build a classifier which can use these parameters to determine the signal strength or quality.

The parameters in the dataset have continous values and therefore their uni, bi and multi-variate EDA's wouldn't have anything substantial to offer for improving the classifier NN model from the Regressor NN model.

```python
df.Signal_Strength.nunique()
```

```python
y_train_cl= to_categorical(y_train)
y_test_cl= to_categorical(y_test)
y_train_cl.shape
```

```python
## Training model-1 for Neural Network Classifier
def classifier():
    model= Sequential()
    model.add(Dense(300,input_shape=(11,)))
    model.add(BatchNormalization())
    model.add(Activation('tanh'))
    model.add(BatchNormalization())
    model.add(Dense(9))
    model.add(Activation('softmax'))
    sgd = optimizers.Adam(lr = 0.001)
    model.compile(optimizer = sgd, loss = 'categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model_one= classifier()
history= model_one.fit(X_train, y_train_cl, epochs=100, batch_size=200, verbose=1, validation_split=0.3 )
```

```python
## Training model-2 for Neural Network Classifier
def classifier_1():
    model = Sequential()
    model.add(Dense(100,input_shape=(11,)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dense(50))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dense(9))
    model.add(Activation('softmax'))
    sgd = optimizers.Adam(lr = 0.001)
   
    model.compile(optimizer = sgd, loss = 'categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model_2= classifier_1()
peak= model_2.fit(X_train, y_train_cl, epochs=100, batch_size=200, verbose=1, validation_split=0.3 )
```

```python
## Training model-3 for Neural Network Classifier
def classifier_3():
    model = Sequential()
    model.add(Dense(100, input_shape=(11,)))
    model.add(BatchNormalization())
    model.add(Activation('tanh'))
    model.add(BatchNormalization())
    model.add(Dense(100))
    model.add(Activation('tanh'))
    model.add(BatchNormalization())
    model.add(Dense(100))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dense(9))
    model.add(Activation('softmax'))
    sgd = optimizers.Adam(lr = 0.001)
   
    model.compile(optimizer = sgd, loss = 'categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model_3= classifier_3()
sigma= model_3.fit(X_train, y_train_cl, epochs=100, batch_size=200, verbose=1, validation_split=0.3)
```
